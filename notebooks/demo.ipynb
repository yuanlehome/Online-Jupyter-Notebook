# Binder Demo Notebook

本示例用于验证 Binder 启动后可运行：
- numpy / pandas 基础操作
- torch 张量计算（CPU）
- transformers 轻量示例（可选）


```python
import numpy as np
import pandas as pd
import torch

print('numpy:', np.__version__)
print('pandas:', pd.__version__)
print('torch:', torch.__version__)
```

```python
# numpy & pandas 基础操作
arr = np.arange(6).reshape(2, 3)
df = pd.DataFrame(arr, columns=['a', 'b', 'c'])
df['sum'] = df.sum(axis=1)
df
```

```python
# torch 张量计算（CPU）
x = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
y = torch.tensor([[10.0, 20.0], [30.0, 40.0]])
z = x @ y
z
```

## 可选：transformers 轻量示例

如果你使用的是默认依赖（`environment.yml`），此单元可直接运行。
如使用自定义最小依赖且未安装 transformers，请跳过。

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('hf-internal-testing/tiny-random-bert')
encoded = tokenizer('Hello Binder!', return_tensors='pt')
encoded
```
